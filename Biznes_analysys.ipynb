```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb



```

```python

df = pd.read_csv("/content/sample_data/sotuvlar.csv")
df.head(20)
```

Asaka filiali

Product name

```python
df_A = df[df.Branch == 'A']
# af = df_A
# af.dropna()
af = df_A.groupby('Product line',as_index=False)#['Quantity'].sum()
# af
# af = df_A.groupby('Product line')
af=af.sum()
af
af.describe()
af.nlargest(10,'Quantity')
plt.figure(figsize=(10,10))
sb.barplot(data=af,x='Quantity',y='Product line')
plt.title('Total Gross Income',fontsize=20)
plt.xlabel('Total Sales',fontsize=18)
plt.ylabel('Product Category Name',fontsize=18)
plt.show()
```
